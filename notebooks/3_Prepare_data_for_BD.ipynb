## Calculate statistics

```python
def statistics_of_tags(textname, text, threshold = 100):
    tags = ['THREAT_ACTOR', 'SOFTWARE', 'INDUSTRY', 'ORG', 'TIMESTAMP',
       'MALWARE', 'COUNTRY', 'IOC', 'IDENTITY', 'CAMPAIGN', 'TOOL',
       'MITRE_ATTACK', 'THEAT_ACTOR', 'ATTACK_PATTERN', 'TECHNIQUE',
       'CITY','CVE']
    
    tags_small = [x.lower() for x in tags]
    
    from bs4 import BeautifulSoup

    from collections import Counter
    soup = BeautifulSoup(text)
    tag_dict = {}
    for tag in tags_small:
        tag_dict[tag] = {}
        kws = []
        for p_or_l in soup.find_all([tag]):
            kws.append(p_or_l.text)
        tag_dict[tag]['KeyWords'] = list(dict(Counter(kws).most_common()[:threshold]).keys())
        tag_dict[tag]['Sum'] = sum(Counter(kws).values())
    raw = dict([('DocName',textname), ('text',text)]+[(tag+'_KeyWords', ';'.join(tag_dict[tag]['KeyWords'])) for tag in tags_small]+\
    [(tag+'_Sum', tag_dict[tag]['Sum'])for tag in tags_small])
    return raw, tag_dict
```

```python
text_stats = []
tags_dicts = []
for i in range(len(parsed_texts)):
    raw,tag_dict = statistics_of_tags(parsed_texts.iloc[i]['DocName'], parsed_texts.iloc[i]['text'])
    text_stats.append(raw)
    tags_dicts.append({'DocName':parsed_texts.iloc[i]['DocName'],'Stats':tag_dict,'Parsed':True})

text_stats = pd.DataFrame(text_stats)

text_stats.head()
```

```python
import json
with open(os.path.join(DATA_DRAFT_PATH, 'parsed_text_stats.json'),'w') as f:
    json.dump(tags_dicts,f)
```

```python
text_stats.to_csv(os.path.join(DATA_DRAFT_PATH, 'parsed_text_stats.csv'), index = False)
```

```python
def load_not_parsed_data(DATA_PATH):
    docs = []
    sent_num_dataset = 0
    for direct in glob.glob(DATA_PATH+'*'):
        for doc in glob.glob(direct+'/*'):
            year = doc.split('/')[-2]
            with open(doc) as f:
                docname = '.'.join(doc.split('/')[-1].split('.')[:-2])
                data = f.read().strip()

            docs.append({'DocName':docname, 'Year':year, 'text':data})
    return docs
```
